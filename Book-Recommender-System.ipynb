### Collaborative filtering approach


```python
x = Ratings_with_bookinfo.groupby('User-ID').count()['Book-Rating'] > 200
educated_users = x[x].index
```

```python
Y = Ratings_with_bookinfo[Ratings_with_bookinfo['User-ID'].isin(educated_users)]
Y
```

```python
X = Y.groupby(by='Book-Title').count()['Book-Rating']>=50
X[X]
```

```python
rated_books = X[X].index
rated_books
```

```python
Final_rating = Y[Y['Book-Title'].isin(rated_books)]
Final_rating
```

```python
table = Final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
table
```

```python
table.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score = cosine_similarity(table)
print(Similarity_score.shape)
Similarity_score
```

```python
def recommend(book_name):
    index =  np.where(table.index == book_name )[0][0]
    similar_books = sorted(list(enumerate(Similarity_score[index])),key=lambda x:x[1],reverse=True)[1:8]
    
    for i in similar_books:
        print(table.index[i[0]])
    
```

```python
recommend('1984')
```
